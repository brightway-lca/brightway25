```python
from scipy.sparse import linalg
from scipy.stats import rankdata
from brightway2 import *
from time import time
import numpy as np
```

http://docs.scipy.org/doc/scipy/reference/sparse.linalg.html#module-scipy.sparse.linalg

```python
solvers = [
    linalg.bicg,
    linalg.bicgstab,
    linalg.cg,
    linalg.cgs,
    linalg.gmres,
    linalg.lgmres,
    # linalg.minres,  # Too slow
    linalg.qmr,
    linalg.spsolve,
]
```

```python
s_indices = {s.__name__: i for i, s in enumerate(solvers)}
```

```python
results = {s.__name__: [] for s in solvers}
```

* bicg(A, b[, x0, tol, maxiter, xtype, M, ...])	Use BIConjugate Gradient iteration to solve A x = b
* bicgstab(A, b[, x0, tol, maxiter, xtype, M, ...])	Use BIConjugate Gradient STABilized iteration to solve A x = b
* cg(A, b[, x0, tol, maxiter, xtype, M, callback])	Use Conjugate Gradient iteration to solve A x = b
* cgs(A, b[, x0, tol, maxiter, xtype, M, callback])	Use Conjugate Gradient Squared iteration to solve A x = b
* gmres(A, b[, x0, tol, restart, maxiter, ...])	Use Generalized Minimal RESidual iteration to solve A x = b.
* lgmres(A, b[, x0, tol, maxiter, M, ...])	Solve a matrix equation using the LGMRES algorithm.
* minres(A, b[, x0, shift, tol, maxiter, ...])	Use MINimum RESidual iteration to solve Ax=b
* qmr(A, b[, x0, tol, maxiter, xtype, M1, M2, ...])

```python
ei = Database("ecoinvent 2.2")
```

```python
method = methods.random()
method
```

```python
def test_all_solvers(mc, results):
    for solver in solvers:
        start = time()
        if solver == linalg.spsolve:
            solver(
                mc.technosphere_matrix,
                mc.demand_array,
            )
        else:
            solver(
                mc.technosphere_matrix,
                mc.demand_array,
                x0=mc.guess,
                maxiter=1000
            )
        results[solver.__name__].append(time() - start)
```

```python
for x in range(100):
    mc = MonteCarloLCA({ei.random(): 1}, method)
    mc.load_data()
    mc.next()
    test_all_solvers(mc, results)
```
